1.   Tipe data nominal merupakan data diskrit dan tidak memiliki urutan. Data nominal hanya berisi atribut atau nama. Seperti variabel cabang dan profesi

#E. Segmentasi menggunakan teknik clustering terhadap RFM

```python
#Recency
#n_hari_terakhir_transaksi sebagai recency

df_recency = pd.DataFrame(df_tp.groupby('cabang')['n_hari_terakhir_transaksi'].max()).reset_index()
df_recency.head()
```

```python
df_recency['n_hari_terakhir_transaksi'] = pd.to_numeric(df_recency['n_hari_terakhir_transaksi'])
df_recency['recency'] =  (pd.to_numeric('500') - df_recency["n_hari_terakhir_transaksi"])
df_recency['recency'] = pd.to_numeric(df_recency['recency'], downcast='integer')
df_recency
```

```python
#Frequency
#jumlah_transaksi sebagai frequency
df_frequency = pd.DataFrame(df_tp.groupby('cabang')['jumlah_transaksi'].count()).reset_index()
df_frequency
```

```python
#Monetary
#nilai_transaksi sebagai monetary
df_tp["monetary"] = (df_tp['nilai_transaksi'] * df_tp['jumlah_transaksi']) \
                                        -  (df_tp['lama_menjadi_customer'] * df_tp['jumlah_transaksi'])
```

```python
df_tp.head()
```

```python
df_monetary = pd.DataFrame(df_tp.groupby('cabang')['monetary'].sum()).reset_index()
```

```python
df_monetary.head(10)
```

```python
fig = make_subplots(rows=1, cols=3, subplot_titles=("Recency", "Frequency", "Monetary"))

fig.add_trace(go.Box(y=df_recency, boxpoints="all"), row=1, col=1)
fig.add_trace(go.Box(y=df_frequency, boxpoints="all"), row=1, col=2)
fig.add_trace(go.Box(y=df_monetary, boxpoints="all"), row=1, col=3)

fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = make_subplots(rows=1, cols=3, subplot_titles=("Recency", "Frequency", "Monetary"))

fig.add_trace(go.Box(y=df_recency['recency'], boxpoints="all"), row=1, col=1)
fig.add_trace(go.Box(y=df_monetary['monetary'], boxpoints="all"), row=1, col=3)

fig.update_layout(showlegend=False)
fig.show()
```

```python
df_RFM = df_recency.merge(df_frequency, left_on="cabang", right_on="cabang").\
                             merge(df_monetary, left_on="cabang", right_on="cabang")
```

```python
df_RFM.head(5)
```

```python
df_RFM.columns = ["cabang",	"hari terakhir transaksi",	"recency",	"frequency", "monetary"]
```

```python
df_RFM.head(5)
```

```python
def get_R_score(recency):
# 0.25    248.0
# 0.50    263.0
# 0.75    298.0

  if recency >= 0 and recency <= 248.0:
    return 1
  elif recency > 248.0 and recency <= 263.0:
    return 2
  elif recency > 263.0 and recency <= 298.0:
    return 3
  elif recency > 298:
    return 4

def get_F_score(frequency):
  # 0.25    11.0
  # 0.50    20.0
  # 0.75    31.0
  if frequency >= 0 and frequency <= 11.0:
    return 1
  elif frequency > 11.0 and frequency <= 20.0:
    return 2
  elif frequency > 20.0 and frequency <= 31.0:
    return 3
  elif frequency > 31:
    return 4


def get_M_score(monetary):
  # 0.25     3361.00
  # 0.50     7553.80
  # 0.75    18136.35
  if monetary >= 0 and monetary <= 3361.0:
    return 1
  elif monetary > 3361.0 and monetary <= 7553.0:
    return 2
  elif monetary > 7553.0 and monetary <= 18136.0:
    return 3
  elif monetary > 18136:
    return 4
```

```python
df_RFM["R_score"] = df_RFM.apply(lambda x:get_R_score(x["recency"]), axis=1)
```

```python
df_RFM["F_score"] = df_RFM.apply(lambda x:get_F_score(x["frequency"]), axis=1)
```

```python
df_RFM["M_score"] = df_RFM.apply(lambda x:get_M_score(x["monetary"]), axis=1)
```
